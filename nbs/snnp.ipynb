# Check Grad

```python
from scipy.optimize import check_grad
```

```python
%run ../layers.py
%run ../criterion.py
%run ../network.py
```

```python
Q = Network()

Q.layers = [Linear(2,6), ReLu(), Dropout(), Linear(6,2), ReLu(), LogSoftMax()]
Q.criterion = NLLCriterion()
```

```python
def func(X):
    X = X.reshape((2,-1))
    Q.train(X,y,0.00)
    return Q.output.sum()
```

```python
def grad(X):
    X = X.reshape((2,-1))
    Q.train(X,y,0.00)
    return Q.layers[0].grad_input.flatten()
```

```python
check_grad(func, grad, X.flatten())
```
